# load dataset

```python
# dataset
import pandas as pd
data = pd.read_excel('features.xlsx')
data.head()
```

# select all vendors

```python
all_vendors=[]
count=[]
for index, row in data.iterrows():
    if index<6000:
        try:
            for i in  row['cvedetail_affected vendor'].split(', '):
                if i not in all_vendors:
                    all_vendors.append(i)
                    count.append(0)
            
        except:
            all_vendors.sort
        
len(all_vendors)
all_vendors.sort()
```

# count vendors

```python
for index, row in data.iterrows():
    if index<6000:
        try:
            for i in  row['cvedetail_affected vendor'].split(', '):
                for j in range(0,len(all_vendors)):
                    if all_vendors[j]==i:
                        count[j]+=1
            
        except:
            for j in range(0,len(all_vendors)):
                if all_vendors[j]=='NA':
                    count[j]+=1
```

```python
from bokeh.plotting import figure, show, output_file,output_notebook
from bokeh.models import ColumnDataSource, Range1d, LabelSet, Label
import numpy

source = ColumnDataSource(data=dict(height=count,
                                    weight=all_vendors,
                                    names=count))
h = numpy.array(count)
df = pd.DataFrame()
df['y_att1'] = count
df['x_att1'] = all_vendors
p = figure(plot_width=1500, plot_height=400,title='Twitter上討論的CVE其Vender統計',x_range=df['x_att1'].values.tolist())
p.rect(x='weight', y=(h/2), width=0.4, height=h,source=source)


p.xaxis[0].axis_label = '廠商(venders)'
p.yaxis[0].axis_label = '數量(count)'

p.left[0].formatter.use_scientific = False
labels = LabelSet(x='weight', y='height', text='names', level='glyph',
              x_offset=-8, y_offset=5, source=source, render_mode='canvas',text_font_size='6pt')
p.y_range.start = 0
p.y_range.end = df['y_att1'].max()*1.1

p.add_layout(labels)
output_notebook()
show(p)
```
