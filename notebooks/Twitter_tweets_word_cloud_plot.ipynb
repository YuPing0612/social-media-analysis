```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt



# Generate a word cloud image
wordcloud1 = WordCloud(width=800, height=400, scale=2,  max_words=2000, min_font_size=8).generate(text)

# show
plt.figure(figsize=(20,10))
plt.imshow(wordcloud1, interpolation='bilinear')
plt.axis("off")
plt.show()

wordcloud2 = WordCloud(width=900,height=500, background_color="white", max_words=2000).generate(text)

# show
plt.figure(figsize=(20,10))
plt.imshow(wordcloud2, interpolation='bilinear')
plt.axis("off")
plt.show()
```
